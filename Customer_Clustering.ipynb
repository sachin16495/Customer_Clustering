```python
import pandas as pd
```

```python
mall_df=pd.read_csv("Mall_Customers.csv")
```

```python
mall_df.head()
```

```python
mall_df=mall_df[['Gender','Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
mall_df.head()
```

```python
dic={'Male':1,'Female':0}
```

```python
mall_df['Gender']=mall_df['Gender'].map(dic)
```

```python
import seaborn as sns
```

```python
sns.pairplot(mall_df)
```

```python
from sklearn.cluster import KMeans
```

```python
customer_cluster1 = KMeans(n_clusters=2, random_state=0).fit(mall_df)
```

```python
mall_df['Cluster 1']=customer_cluster1.labels_
```

```python
sns.pairplot(mall_df)
```

```python
mall_df=mall_df.drop('Cluster 1',axis=1)
```

```python
customer_cluster2 = KMeans(n_clusters=3, random_state=0).fit(mall_df)
```

```python
mall_df['Cluster 2']=customer_cluster2.labels_
```

```python
sns.pairplot(mall_df)
```
